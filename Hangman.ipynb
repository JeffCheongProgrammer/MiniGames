```python
import random
from Hangman_tools import player_lives_display
from Hangman_words import list_A
word_list = list_A
chosen_word = random.choice(word_list)
length_chosen_word = len(chosen_word)
guessed_word = []
is_gameover = False
player_lives = 6
user_guess = ""
user_choice_1 = ""
char = ""
char_index = 0

def display_guessed_word():
    display_word = ""
    for char in guessed_word:
        display_word += char + " "
    print(f'''This is the word that you are guessing:
{display_word}\n''')

def display_player_lives():
    print(f'''There are {player_lives} tries remaining:\n
    {player_lives_display[player_lives]}''')

def player_win():
    print(f'''"{user_guess}" is the correct answer. You won the hangman game!''')

def player_lost():
    print()
    print(f'''You have not guessed the word correctly and ran out of tries.
Game Over:(
{player_lives_display[player_lives]}''')

for char in range(0,length_chosen_word):
    guessed_word.append("_")

#Start game msg:
print(f'''============================
Welcome to the Hangman Game!
============================
{player_lives_display[6]}
The word to be guessed has {length_chosen_word} letters and they are all in lower case. Hint: {chosen_word}''')
display_guessed_word()
print('''You have a total of 6 tries in this game.
Good Luck!\n''')

#Play game:
while not is_gameover:
    user_choice_1 = input('''Do you wish to guess the word or guess a letter within the word?
(Type 'w' to guess the word or type 'l' to guess a letter in the word)''').lower()
    
    if user_choice_1 == "w":
        user_guess = input('''What is the word?''').lower()
        if user_guess == chosen_word:
            player_win()
            is_gameover = True
        else:
            player_lives -= 1
            if player_lives == 0:
                is_gameover = True
                player_lost()
            else:
                print()
                print(f'''The letter "{user_guess}" is not in the word.\n''')
                display_guessed_word()
                display_player_lives()
    
    elif user_choice_1 == "l":
        user_guess = input('''What is the letter?''').lower()
        if user_guess not in chosen_word:
            player_lives -= 1
            if player_lives == 0:
                is_gameover = True
                player_lost()
            else:
                print()
                print(f'''You did not guess the letter correctly.\n''')
                display_guessed_word()
                display_player_lives()
        else:
            for char_index in range(0,length_chosen_word):
                if user_guess == chosen_word[char_index]:
                    guessed_word[char_index] = user_guess
            if "_" not in guessed_word:
                player_win()
                is_gameover = True
            else:
                print(f'''The letter {user_guess} is revealed.''')
                display_guessed_word()            
    else:
        print('''Invalid user input...pls try again:)\n''')
            
    
```
